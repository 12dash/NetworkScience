```python
import pandas as pd
import numpy as np
import os
import xml.etree.ElementTree as ET
```

```python
data=pd.read_csv('Faculty.csv')
data=data.drop(data.columns[[4]], axis=1)

for i in range(2000, 2021):
    data[str(i)]=''

data['Aliases']=''
```

```python
nameid={}

for x in os.listdir():
    if(x[-3:]!='xml'):
        continue
    name=x[:-4]
    
    xml=ET.parse(name+'.xml')
    
    root=xml.getroot()
    
    pid=root.attrib['pid']
    
    nameid[name]=pid
    nameid[pid]=name
    
nameid
```

```python
for x in os.listdir():
    if(x[-3:]!='xml'):
        continue
    name=x[:-4]
    
    print(name)

    xml=ET.parse(name+'.xml')

    root=xml.getroot()
    
    collaborations=[]
    
    for i in range(22):
        collaborations.append([])
        
    for x in root:
        if(x.tag=='r'):
            year=2000
            coauth=[]
            for y in x[0]:
                if((y.tag=='author' or y.tag=='editor') and y.attrib['pid'] in nameid):
                    coauth.append(nameid[y.attrib['pid']])
                if(y.tag=='year'):
                    year=int(y.text)
            index=year-2000
            for x in coauth:
                if(x not in collaborations[index]):
                    collaborations[index].append(x)

    for i in range(len(collaborations)):
        if(name in collaborations[i]):
            collaborations[i].remove(name)
        for j in range(i):
            for k in range(len(collaborations[j])):
                if(collaborations[j][k] not in collaborations[i]):
                    collaborations[i].append(collaborations[j][k])

    for i in range(len(collaborations)):
        year='20'
        if(i<10):
            year+='0'+str(i)
        else:
            year+=str(i)
        
        if(len(collaborations[i])==0):
            data.loc[data['Faculty']==name, year]='[]'
        else:
            data.loc[data['Faculty']==name, year]=str(collaborations[i])
            
    
    
```

```python
data.to_csv('SCSENodes.csv')
```
